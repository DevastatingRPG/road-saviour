```python
import image_processing as ip
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Specify the path to the dataset folder
dataset_folder = "../dataset"

# Get a list of all image files in the dataset folder
image_files = [f for f in os.listdir(dataset_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Check if there are any images in the dataset folder
if not image_files:
    print("No images found in the dataset folder.")
    sys.exit(1)

# Select 10 random image files
random_image_files = np.random.choice(image_files, size=10, replace=False)

# Create a grid for plotting (2 rows x 5 columns)
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()  # Flatten the 2D array of axes for easier iteration

# Process each image and plot
for idx, image_file in enumerate(random_image_files):
    image_path = os.path.join(dataset_folder, image_file)

    # Run signal detection
    signal = ip.signal_detection(image_path)

    # Run zebra detection
    zebra_detected = ip.zebra_detection(image_path)

    # Load the image
    image = cv2.imread("combined.png")
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Plot the image with results
    axes[idx].imshow(image_rgb)
    axes[idx].set_title(f"Signal: {signal}, Zebra: {zebra_detected}")
    axes[idx].axis("off")

# Hide any unused axes
for idx in range(len(random_image_files), len(axes)):
    axes[idx].axis("off")

# Adjust layout and display the grid
plt.tight_layout()
plt.show()
```
